```python
class Web_link(BaseModel):
    title:str = Field(description="链接标题")
    link: str = Field(
        description="网页超链接"
    )
    date: str = Field(None,description="链接发布日期")

class Link_Collection(BaseModel):
    collections:list[Web_link] = Field(None,description="链接集合")
```

```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model=model,temperature=0,max_tokens=None,timeout=None,max_retries=2,api_key=api_key,base_url=base_url)
```

```python
structured_llm = llm.with_structured_output(Link_Collection)
```

```python
messages = [
    (
        "system",
        "根据用户提供的信息进行数据抽取，严禁遗漏和虚构信息。",
    ),
    ("human", raw_markdown),
]
```

```python
result = structured_llm.invoke(messages)
```
