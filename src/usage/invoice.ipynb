````python
from concurrent.futures import ThreadPoolExecutor
from typing import List, Dict, Callable
````

````python
def parallel(messages: List, n_workers: int = 3) -> List[str]:
    """Process multiple inputs concurrently with the same prompt."""
    with ThreadPoolExecutor(max_workers=n_workers) as executor:
        futures = [executor.submit(chat_with_llm, message=x, api_key=api_key, base_url=base_url,model_name='/root/Qwen2.5-VL-32B-Instruct-AWQ', temperature=0.9) for x in messages]
        return [f.result() for f in futures]
````
